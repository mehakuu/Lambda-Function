```python
sqaure = lambda x : x*x
sqaure(6)
```

```python
sum1 = lambda a,b,c: a+b+c
sum1(1,2,3)
```

```python
l =[2,3,4]
print(list(map(sqaure,l))) # map will return sqaure of all the elements of l list 
```

```python
from functools import reduce

sum = lambda a, b: a+b

l = [1, 2, 3, 4]
val = reduce(sum, l)
print(val)
```

```python
g10 = lambda num: num>10

l = [1, 2, 3, 4, 5, 6, 7, 8, 89, 98]
print(list(filter(g10, l)))

# filter out those values which are greater than 10 
```
